```python
import tkinter as tk
import datetime
import requests
import json
```

```python
#call API
response = requests.get("https://api.open-meteo.com/v1/forecast?latitude=-33.9249&longitude=18.4241&current_weather=true&hourly=temperature_2m,windspeed_10m,winddirection_10m,weathercode")
```

```python
#print API response
text = json.dumps(response.json(), sort_keys=True, indent=4)
print(text)
d = response.json()
```

```python
current_temp = d["current_weather"]["temperature"]
current_time = d["current_weather"]["time"]
time = datetime.datetime.fromisoformat(current_time)

#time = time.strftime("%H")
current_winddirection = d["current_weather"]["winddirection"]
current_windspeed = d["current_weather"]["windspeed"]
print("Current temperature:",current_temp)
print(time)
```
